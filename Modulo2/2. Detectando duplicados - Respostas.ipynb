- Adesão marginal: 1-10
- Tamanho de célula epitelial única: 1–10
- Núcleos nus: 1-10

#### 3. Jogue fora os dados ausentes

```python
df.dropna(axis = 0, how = 'any', inplace=True)
```

#### 4. Use a função `duplicated()` para checar quais são os dados duplicados

```python
df[df.duplicated(keep=False)]
```

Agora que lidamos com linhas inteiras que foram duplicadas, também precisamos lidar com pacientes duplicados. Lembre-se, se estivermos usando um modelo de ML para isso, precisamos nos certificar de que nenhum paciente seja repetido mais do que a média de vezes que um paciente é repetido, para que nosso modelo de ML não seja tendencioso.
